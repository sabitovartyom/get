```python
import os
```

```python
names = []
with os.scandir('./CS_SEPMP_2023-main/Seminars_2/Sem1_dimers_data') as entries:
    for entry in entries:
        name = entry.name
        names.append(name)

data = []
for i in range(len(names)):
    with open('./CS_SEPMP_2023-main/Seminars_2/Sem1_dimers_data/' + names[i], 'r') as f:
        dt = f.readlines()
        if i == 0:
            columns = ['File'] + dt[57].split()
        try:
            dt = dt[58].split()
            data.append([names[i]] + dt)
        except IndexError:
            None
        f.close()

import pandas as pd
df = pd.DataFrame(data, columns=columns)
df.to_csv('./graphen.csv')
df
```

```python
df2 = pd.read_csv('./graphen.csv')
columns2 = columns[1:]
summ = []
mean = []
standard_deviation = []

for i in columns2:
    summ.append(df2[i].sum())
    mean.append(df2[i].mean())
    standard_deviation.append(df2[i].std())

data2 = {'параметр': columns2, 'сумма':summ, 'среднее':mean, 'стандартное отклонение':standard_deviation}
df3 = pd.DataFrame(data2)
df3
```
